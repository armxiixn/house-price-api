```python
import pandas as pd
import re




pattern = re.compile(r'[^A-Za-z0-9آ-ی\s]')

def clean_with_regex(text):
    if pd.isnull(text):
        return text
    text = str(text)
    cleaned = pattern.sub('', text)          
    cleaned = re.sub(r'\s+', ' ', cleaned)   
    return cleaned.strip()

df = df.applymap(clean_with_regex)

df.to_csv('housePrice.csv', index=False)
```

```python

import pandas as pd

df['Area'] = (
    df['Area']
        .astype(str)
        .str.replace(r'[^\d]', '', regex=True)
        .replace('', pd.NA)                     
)
df['Area'] = pd.to_numeric(df['Area'], errors='coerce').astype('int64')
df.info()
```

```python
from sklearn.preprocessing import StandardScaler, LabelEncoder 
le = LabelEncoder()
for column in ['Elevator','Parking','Warehouse']:
    df[column] = le.fit_transform(df[column])   
df    
```

```python
df.info()
```

```python


cols_to_convert = ["Room","Price"  ]


for col in cols_to_convert:
    df[col] = pd.to_numeric(df[col]).astype(int)

print(df.dtypes) 
```

```python

print(df.isnull().sum())
```

```python
Q1 = df['Price'].quantile(0.25)
Q3 = df['Price'].quantile(0.75)
IQR = Q3 - Q1

low_limit = Q1 - 1.5 * IQR
high_limit = Q3 + 1.5 * IQR
print("Before:", df.shape)

df = df[(df['Price'] >= low_limit) & (df['Price'] <= high_limit)].copy()


print("After:", df.shape)
```

```python

df['Area_log'] = np.log1p(df['Area'])


Q1 = df['Area_log'].quantile(0.25)
Q3 = df['Area_log'].quantile(0.75)
IQR = Q3 - Q1
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

df = df[(df['Area_log'] >= lower) & (df['Area_log'] <= upper)]
```

```python
df.shape
```

```python

df['majmooe_emkanat'] = df['Parking'] + df['Warehouse'] + df['Elevator']
df['Area_per_Room'] = df['Area'] / df['Room']
df['nesbat_otagh_be_area'] = df['Room'] / df['Area']


```

```python
df
```

```python
df.replace([np.inf, -np.inf], np.nan, inplace=True)
df['Area_per_Room'].fillna(df['Area_per_Room'].median(), inplace=True)
df['nesbat_otagh_be_area'].fillna(df['nesbat_otagh_be_area'].median(), inplace=True)
```

```python
df['Price'] = np.log1p(df['Price'])

```

```python
region_mean_price = df.groupby('Address')['Price'].mean()
df['Region_mean_price'] = df['Address'].map(region_mean_price)
```

```python
df['Region_mean_price'] = df['Region_mean_price'].fillna(df['Region_mean_price'].mean())
```
